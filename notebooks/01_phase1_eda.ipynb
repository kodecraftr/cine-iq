Run `preprocess.py` first, then use this notebook to verify outputs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

movies   = pd.read_csv('../../data/processed/movies_merged.csv')
ratings  = pd.read_csv('../../data/processed/ratings_clean.csv')

print('Movies shape :', movies.shape)
print('Ratings shape:', ratings.shape)
movies.head(3)
```

```python
# Missing values
print(movies.isnull().sum())
```

```python
# Rating distribution
ratings['rating'].hist(bins=10)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Top genres
from collections import Counter
all_genres = [g for genres in movies['genres'].dropna() for g in ast.literal_eval(genres)]
genre_counts = Counter(all_genres).most_common(15)
labels, values = zip(*genre_counts)
plt.barh(labels[::-1], values[::-1])
plt.title('Top 15 Genres')
plt.tight_layout()
plt.show()
```

```python
# Movies per decade
movies['decade'] = (movies['release_year'] // 10 * 10).astype('Int64')
movies.groupby('decade').size().plot(kind='bar')
plt.title('Movies per Decade')
plt.tight_layout()
plt.show()
```

```python
# Ratings per user (sparsity check)
ratings_per_user = ratings.groupby('userId').size()
print(ratings_per_user.describe())
ratings_per_user.clip(upper=200).hist(bins=40)
plt.title('Ratings per User (clipped at 200)')
plt.show()
```
